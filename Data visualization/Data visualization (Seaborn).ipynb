```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# path of the file to read
file_path = './fifa.csv'
fifa_data = pd.read_csv(file_path, index_col='Date', parse_dates=True)
fifa_data.head()
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=fifa_data)
```

```python
spotify_file_path = './spotify.csv'
spotify_data= pd.read_csv(spotify_file_path, index_col='Date', parse_dates=True)

# spotify_data.head()

# set the width and height of the figure
plt.figure(figsize=(14, 6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

sns.lineplot(data=spotify_data)
```

```python
flight_filepath = './flight_delays.csv'

flight_data = pd.read_csv(flight_filepath, index_col='Month')
flight_data
```

```python
plt.figure(figsize=(10, 6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap(data=flight_data, annot=True)
plt.xlabel("Airline")
```

```python
insurance_filepath = "./insurance.csv"

insurance_data = pd.read_csv(insurance_filepath)
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# with line of regression (best-fit)
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

```python
iris_filepath = './iris.csv'

iris_data = pd.read_csv(iris_filepath, index_col='Id')
iris_data.head()
```

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde')
```

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

plt.title("Histogram of Petal Lengths, by Species")
```

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

plt.title("Distribution of Petal Lengths, by Species")
```
